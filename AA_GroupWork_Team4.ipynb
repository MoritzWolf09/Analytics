```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

nextbike_leipzig = pd.read_csv('dataset/nextbike/leipzig.csv', 
                           delimiter=',')
nextbike_leipzig['Datetime'] = nextbike_leipzig[['day', 'time']].apply(lambda x: ''.join(x), axis=1).str.replace('-', '').str[:10]
nextbike_leipzig.sort_values(by=['Datetime'])
```

```python
nextbike_count = nextbike_leipzig.groupby('Datetime').count()[['b_number']]
nextbike_count
```

```python
niederschlag_leipzig = pd.read_csv('dataset/wetterdaten/leipzig/produkt_rr_stunde_20190608_20201208_02928.txt', 
                                   delimiter=';',
                                   names = ['stationsnummer','Datetime', 'event', 'niederschlagsmenge', 'indikator niederschlag', 'niederschlagsnummer', 'topic'])

niederschlag_leipzig[['Datetime', 'niederschlagsmenge', 'indikator niederschlag']]
niederschlag_leipzig.sort_values(by=['Datetime'])
```

```python
infos_merged = nextbike_leipzig.merge(niederschlag_leipzig, how = 'left')
infos_merged.groupby(['Datetime', 'indikator niederschlag']).count()[['b_number']]
```

```python
y = infos_merged.groupby(['Datetime', 'indikator niederschlag']).count()[['b_number']]['b_number']
x = infos_merged.groupby(['Datetime', 'indikator niederschlag']).count()[['b_number']]['Datetime']

plt.figure(figsize = (8,6))
plt.scatter(x, y)
plt.xlabel("Datum")
plt.ylabel("Anzahl Ausleihen")
plt.title("Petal length per iris flower")
plt.show()
```
